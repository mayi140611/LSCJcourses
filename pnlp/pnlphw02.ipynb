### 22.◑找出聊天语料库（text5）中所有四个字母的词。使用频率分布函数（FreqDist），以频率从高到低显示这些词。

```python
l = [word for word in text5 if len(word)==4]
fdist = FreqDist(l)
fdist.most_common()
```

### 28.◑定义一个函数percent(word, text)，计算一个给定的词在文本中出现的频率，结果以百分比表示。

```python
def percent(word, text):
    return '{:.2f}%'.format(100*text.count(word)/len(text))

percent('is', text1)
```
